## Data preprocessing

```python
numerical_columns = df.select_dtypes(include='number').columns

print(numerical_columns)
```

```python
# Changing all Yes/No answers to binary, yes = 1, no = 0
yes_no_cols = ['PhysicalActivities', 'HadHeartAttack', 'HadAngina',
       'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD',
       'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
       'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
       'DifficultyConcentrating', 'DifficultyWalking',
       'DifficultyDressingBathing', 'DifficultyErrands',
       'ChestScan', 'AlcoholDrinkers','HIVTesting', 'FluVaxLast12', 'PneumoVaxEver',
       'HighRiskLastYear']

for i in yes_no_cols:
    df[f"{i}"] = df[f"{i}"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
# Changing gender to binary, male = 1, female = 0
df["Sex"] = df["Sex"].apply(lambda x: 1 if x == "Male" else 0)
```

```python
all_cols = ['State', 'Sex', 'GeneralHealth', 'PhysicalHealthDays',
       'MentalHealthDays', 'LastCheckupTime', 'PhysicalActivities',
       'SleepHours', 'RemovedTeeth', 'HadHeartAttack', 'HadAngina',
       'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD',
       'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
       'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
       'DifficultyConcentrating', 'DifficultyWalking',
       'DifficultyDressingBathing', 'DifficultyErrands', 'SmokerStatus',
       'ECigaretteUsage', 'ChestScan', 'RaceEthnicityCategory', 'AgeCategory',
       'HeightInMeters', 'WeightInKilograms', 'BMI', 'AlcoholDrinkers',
       'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'TetanusLast10Tdap',
       'HighRiskLastYear', 'CovidPos'] 

# Yes/No cols, sex
values_to_remove = ['Sex','PhysicalActivities', 'HadHeartAttack', 'HadAngina',
       'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD',
       'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
       'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
       'DifficultyConcentrating', 'DifficultyWalking',
       'DifficultyDressingBathing', 'DifficultyErrands',
       'ChestScan', 'AlcoholDrinkers','HIVTesting', 'FluVaxLast12', 'PneumoVaxEver',
       'HighRiskLastYear']

rest_cols = [x for x in all_cols if x not in values_to_remove]

print(f"Columns not binary: {rest_cols}")

rest_cols = [x for x in rest_cols if x not in numerical_columns]

print(f"Columns not binary nor numerical: {rest_cols}")
```

```python
df.HadDiabetes
```

```python
# Replacing the GeneralHealth values with numerical values

GeneralHealth_order = ["Excellent", "Very good", "Good", "Fair", "Poor"]
GeneralHealth_vals = [5, 4, 3, 2, 1]

GeneralHealth_dict = dict(zip(GeneralHealth_order, GeneralHealth_vals))

df["GeneralHealth"].replace(GeneralHealth_dict, inplace=True)
```

```python
# Setting State to binary
df = pd.concat([df.drop("State", axis=1), pd.get_dummies(df.State, prefix="State_", dtype=int)], axis=1)

# Setting LastCheckupTime to binary
df = pd.concat([df.drop("LastCheckupTime", axis=1), pd.get_dummies(df.LastCheckupTime, prefix="LastCheckupTime_", dtype=int)], axis=1)

# Setting RemovedTeeth to binary
df = pd.concat([df.drop("RemovedTeeth", axis=1), pd.get_dummies(df.RemovedTeeth, prefix="RemovedTeeth_", dtype=int)], axis=1)

# Setting SmokerStatus to binary
df = pd.concat([df.drop("SmokerStatus", axis=1), pd.get_dummies(df.SmokerStatus, prefix="SmokerStatus_", dtype=int)], axis=1)

# Setting RaceEthnicityCategory to binary
df = pd.concat([df.drop("RaceEthnicityCategory", axis=1), pd.get_dummies(df.RaceEthnicityCategory, prefix="RaceEthnicityCategory_", dtype=int)], axis=1)

# Setting AgeCategory to binary
df = pd.concat([df.drop("AgeCategory", axis=1), pd.get_dummies(df.AgeCategory, prefix="AgeCategory_", dtype=int)], axis=1)

# Setting ECigaretteUsage to binary
df = pd.concat([df.drop("ECigaretteUsage", axis=1), pd.get_dummies(df.ECigaretteUsage, prefix="ECigaretteUsage_", dtype=int)], axis=1)

# Setting TetanusLast10Tdap to binary
df = pd.concat([df.drop("TetanusLast10Tdap", axis=1), pd.get_dummies(df.TetanusLast10Tdap, prefix="TetanusLast10Tdap_", dtype=int)], axis=1)

# Setting CovidPos to binary
df = pd.concat([df.drop("CovidPos", axis=1), pd.get_dummies(df.CovidPos, prefix="CovidPos_", dtype=int)], axis=1)

# Setting HadDiabetes to binary
df = pd.concat([df.drop("HadDiabetes", axis=1), pd.get_dummies(df.HadDiabetes, prefix="HadDiabetes_", dtype=int)], axis=1)

df
```

```python
numerical_columns = df.select_dtypes(include='number').columns

print(numerical_columns)
```

```python
non_numerical_columns = df.select_dtypes(exclude='number').columns
print(non_numerical_columns)
```

```python
# Drop columns that start with "State_"
filtered_df = df.drop(columns=[col for col in df.columns if col.startswith("State_")])

# Calculate the correlation matrix
corr_matrix = filtered_df.corr()

# Plot the heatmap with filtered correlations
plt.figure(figsize=(24, 20))
sns.heatmap(corr_matrix, annot=False, cmap="coolwarm")
plt.show()
```
